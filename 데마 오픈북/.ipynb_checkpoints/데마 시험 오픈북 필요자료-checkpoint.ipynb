## 1번

```python
import plotly
import plotly.express as px
import plotly.graph_objs as go 
geojson = px.data.election_geojson()
df= plotly.data.election()
df
```

#### 1-1: Joly의 득표률(%)과 상위 5개 데이터

```python
df['JolyPct'] = df['Joly'] / df['total']
df.head()
```

#### 1-2: id가 190 이상인 지역에 대해 득표 비율

```python
select = df.copy()
select = select[select['district_id'] >= 190]
select['CorPct'] = select['Coderre'] / select['total']
select['BerPct'] = select['Bergeron'] / select['total']
fig = px.bar(select, x = 'district', y = ['JolyPct', 'CorPct', 'BerPct'])
fig.show()
df
```

#### 1-3: 1-1을 지도시각화

```python
lp = px.choropleth_mapbox(df, geojson = geojson, zoom = 3, locations = 'JolyPct')
lp.show()
```

## 2번

```python
import seaborn as sns
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

flgith = sns.load_dataset('flights')
flgith
```

#### 2-1: 연도별 누적 승객

```python
num2_1 = flgith.groupby('year')['passengers'].sum()
num2_1
```

#### 2-2: 최대승객 - 최소승객

```python
def f(arr):
    return arr.max() - arr.min()
num2_2 = flgith.groupby('year')['passengers'].agg(f)
num2_2
```

#### 2-3: 1,2 그래프로

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize = (20,7))
plt.subplot(1,2,1)
plt.bar(np.array(num2_1.index), num2_1)
plt.xticks(num2_1.index)
plt.xlabel("year")
plt.ylabel("passenger")

plt.subplot(1,2,2)
plt.bar(np.array(num2_2.index), num2_2)
plt.xticks(num2_2.index)
plt.xlabel("year")
plt.ylabel("max_pass - min_pass")
plt.show()
```

#### 3-2: 회사명 워드클라우드

```python
from wordcloud import WordCloud
txt = ""
for x in Mpg_data['name']:
    txt += x.split()[0]
#     print(txt)
    txt += " "
#     print(x,txt)
wc = WordCloud().generate(txt)

plt.figure(figsize = (12,12))
plt.imshow(wc, interpolation = 'lanczos')
plt.axis('off')
plt.show()
# print(txt)
```

## 4번

```python
import seaborn as sns
penguins = sns.load_dataset('penguins')
penguins
```

#### 4-1: 종별로 나눈 후 island로 분석

```python
Adelie = penguins[penguins['species'] == 'Adelie']
Adelie.pivot_table(index = 'island')
```

- Adelie는 세 지역에 살고있다. (Biscoe, Dream, Torgesen)
- Biscoe에서는 Gentoo, Dream에서는 Chinstrap의 종과 함께 존재하지만 Torgerson에는 하나의 종만 있다.

```python
Chinstrap = penguins[penguins['species'] == 'Chinstrap']
Chinstrap.pivot_table(index = 'island')
```

- Chinstrap은 하나의 지역에 살고있다. (Dream)

```python
Gentoo = penguins[penguins['species'] == 'Gentoo']
Gentoo.pivot_table(index = 'island')
```

- Gentoo는 하나의 지역에 살고있다. (Biscoe)

#### 4-2: 그래프 분석

```python
sns.violinplot(data = penguins, x = penguins['body_mass_g'] // 1000, y = 'flipper_length_mm', 
               hue = 'sex', palette=["lightblue", "lightpink"], split = True)
```

- body_mass_g가 클수록 flipper_length_mm도 크다.
- body_mass_g가 2천대인 성별은 여자밖에 없다.
- body_mass_g가 6천대인 성별은 남자밖에 없다.
- body_mass_g가 4천대인 성별에서는 여자가 flipper_length_mm가 더 길다.
- body_mass_g가 5천대인 성별에서는 남자가 flipper_length_mm가 더 길다.
