2-1 연도별 누적 승객

```python
flgith = sns.load_dataset('flights')
flgith
yearpass = flgith.groupby('year')['passengers'].sum()
yearpass
```

```python
flgith
```

2-2

```python
def f(arr):
    return arr.max() - arr.min()

pass_sub = flgith.groupby('year')['passengers'].agg(f)
pass_sub
```

2-3 연도의 연도 표시

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
plt.bar(np.array(yearpass.index), yearpass)
plt.xticks(yearpass.index)
plt.xlabel("year")
plt.ylabel("passenger")

plt.subplot(1,2,2)
plt.bar(np.array(pass_sub.index), pass_sub)
plt.xticks(pass_sub.index)
plt.xlabel("year")
plt.ylabel("max_pass - min_pass")
plt.show()
```
