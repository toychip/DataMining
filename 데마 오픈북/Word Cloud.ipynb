```python
import numpy as np
from PIL import Image

text = open('./worldcloud/alice.txt').read()  
alice_mask= np.array(Image.open('./worldcloud/alice.png'))  #엘리스의 이미지 모양대로 만들겠다

stopword=set(STOPWORDS)  #금지어 설정
stopword.add('said')
```

```python
plt.figure(figsize=(8,8))
plt.imshow(alice_mask,cmap=plt.cm.gray, interpolation='bilinear')  #이미지만 그냥 출력해봄(어떤 모양인지 보려고)
plt.axis('off')
plt.show
```

```python
wc = WordCloud(background_color='white',max_words=2000,mask=alice_mask, stopwords= stopword)
wc = wc.generate(text)
#max_words: 이미지에 넣을 최대 단어 수
#mask: 그림을 그릴 모양
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import numpy as np
from PIL import Image

#text = open('./data/alice.txt').read()
alice_coloring= np.array(Image.open('./worldcloud/alice_color.png'))

stopword=set(STOPWORDS)
stopword.add('said')

wc = WordCloud(background_color='white',max_words=2000,mask=alice_coloring, stopwords= stopwords)
wc = wc.generate(text)

#색을 지정하지 않으면 이렇게 생김!
plt.figure(figsize=(12,12))
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from wordcloud import ImageColorGenerator
image_colors=ImageColorGenerator(alice_coloring)  #이 사진의 색을 비슷하게 생성
#image_colors.default_color = [0,0,0]
print(image_colors.image.shape)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(alice_coloring,cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show
#이 그림의 색과 유사하게 월드클라우드를 만들기 위함!
```

```python
plt.figure(figsize=(12,12))
#wc의 색을 다시 칠해주는 것이기 때문에 recolor
plt.imshow(wc.recolor(color_func = image_colors), interpolation='bilinear')
plt.axis('off')
plt.show()
```
