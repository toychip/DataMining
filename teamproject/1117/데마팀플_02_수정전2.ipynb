```python
datas =[]
for i in df3:
  datas.extend(i.split('->'))

new_datas=[j.replace(' ', '') for j in datas]
new_datas
```

```python
w_count= {}
for lst in new_datas:
    try: w_count[lst]+= 1
    except: w_count[lst]=1
print(w_count)

# for lst in len(w_count):
#     try: w_count[lst]+= 1
#     except: w_count[lst]=1
```
