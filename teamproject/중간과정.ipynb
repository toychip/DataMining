```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['axes.unicode_minus'] = False
import numpy as np
import seaborn as sns
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

df = pd.read_excel('5_패스생성정보.xlsx')                 # 파일 읽어옴
df.columns = ['패스명', '패스메모', 'road', '등록일시']      # 경로 split하기 위해서 column이름 바꿈
road_list = df.road.str.split(' ->').str[0].tolist()   # ' ->'로 split 후 리스트에 넣음


df = pd.DataFrame(road_list)      # 리스트를 dataframe형식으로 바꿈
                              # 같은것끼리 카운트
# print(df)
df = df.dropna(axis=0)
print(df)
df = df.value_counts()   
df
# df.to_excel('exam.xlsx')

plt.figure(figsize=(25, 10))

p = df[df>=10]
# plt.style.use('t')
plt.title('방문', fontsize=30)
p.plot.bar(width=0.5, fontsize=15)

plt.figure(figsize=(20, 20))
p = df[df>=25]
p.plot.pie(fontsize=10, startangle = 45, counterclock = True)

```
