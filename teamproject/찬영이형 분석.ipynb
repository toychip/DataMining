```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
from plotly.subplots import make_subplots
from sklearn.preprocessing import LabelEncoder
import datetime
import matplotlib.pyplot as plt
import pandas as pd

import numpy as np

data = pd.read_excel('data.xlsx')
```

```python
data
```

```python
data1 = data['1뎁스']
data2 = data['2뎁스']

df = pd.concat([data1, data2], axis=1)
df = df.dropna()
df, type(df)
```

```python
def plot_func(df,category):
    temp_df = df.loc[df['1뎁스']==category]
    temp_df.rename(columns = {'2뎁스':category + '종류'}, inplace = True)
    temp_df_count = temp_df.groupby([category + '종류']).count()
    temp_df_count.rename(columns = {'1뎁스':'가게 수'}, index = {'2뎁스':category + '종류'},inplace = True)
    fig = px.scatter(temp_df_count)
    fig.show()
```

```python
df['1뎁스'].unique()
```

```python
for i in df['1뎁스'].unique():
    plot_func(df, i)
```
