인천관광명소 데이터

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_path = "C:\Windows\Fonts\malgun.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font) ##폰트 한글깨짐 해결

#print(plt.style.available) 다양한 스타일

incheonSight = pd.read_csv('./인천관광명소instaCrawling.csv')
incheonSight
```

```python
plt.figure(figsize=(10, 15))
place = incheonSight['place'].value_counts()
p = place[place>=3]
plt.style.use('fivethirtyeight')
plt.title('인천관광명소')
p.plot.barh(width=0.8, color='skyblue')
```
