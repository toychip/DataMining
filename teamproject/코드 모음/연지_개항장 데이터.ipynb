```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_path = "C:\Windows\Fonts\malgun.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font) ##폰트 한글깨짐 해결

#print(plt.style.available) 다양한 스타일

incheon = pd.read_csv('./개항장 instaCrawling.csv')
incheon
```

```python
plt.figure(figsize=(8, 15))
place = incheon['place'].value_counts()
p = place[place>=5]
plt.style.use('fivethirtyeight')
plt.title('인천개항장', fontsize=18)
p.plot.barh(width=0.8, fontsize=15, color='skyblue')
```

```python
plt.figure(figsize=(20, 20))
place = incheon['place'].value_counts()
p = place[place>=10]
plt.title('인천개항장')
p.plot.pie(fontsize=15)
```
