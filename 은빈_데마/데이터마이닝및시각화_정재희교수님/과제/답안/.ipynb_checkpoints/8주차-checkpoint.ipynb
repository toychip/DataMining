```python
import requests
from pandas import DataFrame
from bs4 import BeautifulSoup
import re
from datetime import datetime
import os

date = str(datetime.now())
date = date[:date.rfind(':')].replace(' ', '_')
date = date.replace(':','시') + '분'
```

```python



query =  '학교'
news_num = 100 
query = query.replace(' ', '+')


news_url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query={}'

req = requests.get(news_url.format(query))
soup = BeautifulSoup(req.text, 'html.parser')


news_dict = {}
idx = 0
cur_page = 1

print()
print('크롤링 중...')
while idx < news_num:
### 네이버 뉴스 웹페이지 구성이 바뀌어 태그명, class 속성 값 등을 수정함(20210126) ###
    table = soup.find('ul',{'class' : 'list_news'})
    li_list = table.find_all('li', {'id': re.compile('sp_nws.*')})
    area_list = [li.find('div', {'class' : 'news_area'}) for li in li_list]
    a_list = [area.find('a', {'class' : 'news_tit'}) for area in area_list]
    for n in a_list[:min(len(a_list), news_num-idx)]:
        news_dict[idx] = {'title' : n.get('title'),
                          }
        idx += 1
    cur_page += 1
    pages = soup.find('div', {'class' : 'sc_page_inner'})
    next_page_url = [p for p in pages.find_all('a') if p.text == str(cur_page)][0].get('href')
    req = requests.get('https://search.naver.com/search.naver' + next_page_url)
    soup = BeautifulSoup(req.text, 'html.parser')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

a = list(news_dict.values())
text = ""
for lst in a:
    text += lst['title']

wc = WordCloud(font_path = 'C:\Windows\Fonts\\malgun.ttf', background_color = 'white').generate(text)

plt.figure(figsize = (11,11))  #그림 크기
plt.imshow(wc, interpolation = 'lanczos')  #이미지 보이기
plt.axis('off')  #축 삭제
plt.show()
```

```python
import numpy as np
import pandas as pd
data = pd.read_csv('nba.csv')
```

```python
data.pivot_table(['Age', 'Salary'], index = ['Position', 'Team'])
```

```python
plt.figure(figsize = (20, 10))
plt.subplots_adjust(hspace = 0.2)

plt.subplot(2, 1, 1)
plt.title("Maximum Salary for each team")
rich = data['Salary'].groupby(data['Team']).max()

lstName = []
for i in range(len(rich)):
    lstName.append(rich.index[i].split()[0][0] + rich.index[i].split()[1][0])
    #split하면 이름과 성이 나뉨, 거기서 첫 글자씩만 가져와서 더한 문자열을 리스트에 추가

plt.bar(lstName, rich)

plt.subplot(2, 1, 2)
plt.title("Maximum Age for each team")
people = data['Age'].groupby(data['Team']).max()
plt.bar(lstName, people)

plt.show()
```
