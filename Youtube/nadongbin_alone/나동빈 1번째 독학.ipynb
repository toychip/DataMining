```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0, np.pi * 10, 500)
fig, axes = plt.subplots(2, 1)            # subplots -> 2개의 그래프가 들어가는 Figure 생성
                                          # 지금은 2개의 행에 1개의 열 그래프
axes[0].plot(x, np.sin(x))
axes[1].plot(x, np.cos(x))


# fig.savefig("sin&cos.png")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-9, 10)
y = x ** 2

# 라인 스타일로는 '-', ':', 'ㅡ.', '--'등이 사용 가능

plt.plot(
    x, y, linestyle = ":", marker = "*",
    markersize=8, markerfacecolor = "blue", markeredgecolor = "red"
)
                            #   marker는 각 점을 해당하는 것.
                            #   linestyle = ":" 이것으로 하면 실선 스타일로 선을 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(10)
y = np.random.rand(10)
z = np.random.rand(10)
data = [x, y, z]
x_array = np.arange(10)
for i in range(0, 3):
    plt.bar(
    x_array,
    data[i],
    bottom = np.sum(data[:i], axis = 0)
    )
plt.show()
```
