```python
import numpy as np
import matplotlib.pyplot as plt

def main():
    Score = np.random.randint(100, size = (10, 4))
    q1_scatter(Score)
    q2_subplot(Score)
    q3_pie(Score)
    q4_boxplot(Score)
    print(Score)
```

```python
def q1_scatter(arr):
    print("1번")
    plt.scatter(arr[:,1], arr[:,0], marker='d')
    plt.xlabel("English")
    plt.ylabel("Korean")
    plt.show()
    print()
```

```python
def q2_subplot(arr):
    print("2번")
    sub = np.array(['Korean', 'English', 'Math', 'Science'])
    for i in range(1, 5):
        plt.subplot(2, 2, i)
        plt.hist(arr[:,i-1])
        plt.xlabel(sub[i-1])
        plt.ylabel('Count')
    plt.show()
    print()
```

```python
def q3_pie(arr):
    print("3번")
    arr = arr[:,0]
    labels = 'A', 'B', 'C', 'D'
    sizes = [(arr>=90).sum(), (arr>=80).sum()-(arr>=90).sum(), (arr>=70).sum()-(arr>=80).sum(), (arr>=60).sum()-(arr>=70).sum()]
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, labels = labels, shadow = True, startangle = 90)
    ax1.axis('equal')
    plt.show()
    print()
```

```python
def q4_boxplot(arr):
    print("4번")
    plt.boxplot(arr, labels = ['Kor', 'Eng', 'Math', 'Science'])
    plt.show()
    print()
```

```python
main()
```
