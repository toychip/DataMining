```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

iris = load_iris()

# 데이터 만들기
d = iris.data  #iris 데이터와 부류를 각각 x, y에 저장
t = iris.target
df = pd.DataFrame(d, columns = iris.feature_names)  # data의 정보 d를 DataFrame형식으로 바꿈  
df["target"] = t.astype('float') # target을 iristarget에 넣었던 t정보를 float형식으로 추가
df["targetName"] =  np.where(t == 0, iris.target_names[0], 
np.where(t == 1, iris.target_names[1], iris.target_names[2]))
# where 절을 이용하여 0이면 0번째, 아니라면 한번더 y가 1인지 확인, 맞으면 1번째를, 아니라면 2번째로 값 추가
df.head()  # 제일 위 5개의 값만 출력
```

```python
# sepal 그래프

setosa = df[df["targetName"] == iris.target_names[0]]
versicolor = df[df["targetName"] == iris.target_names[1]]
virginica = df[df["targetName"] == iris.target_names[2]] 
# bollean 값으로 각 해당하는 targetname의 데이터를 저장

plt.plot(setosa.iloc[:,[0]], setosa.iloc[:,[1]], 'ro', label = "setosa")
# setosa의 전체 데이터의 전체 중 0번째 인덱스(sepal length(cm))와 
# 전체 중 1번째 인덱스(sepal width) 를 통해 x, y축에 해당하는 값을 빨간 원으로 표시, 라벨은 setosa로 표시

plt.plot(versicolor.iloc[:,[0]], versicolor.iloc[:,[1]], 'bx', label = "versicolor") 
# versicolor의 전체 데이터의 전체 중 0번째 인덱스(sepal length(cm)와 
# 전체 중 1번째 인덱스(sepal width) 를 통해 x, y축에 해당하는 값을 파란 x로 표시, 라벨은 versicolor로 푝시

plt.plot(virginica.iloc[:,[0]], virginica.iloc[:,[1]], 'g^', label = "virginica")  
# virginica의 전체 데이터의 전체 중 0번째 인덱스(sepal length(cm)와 
# 1번째 인덱스(sepal width)를 통해 x, y축에 해당하는 값을 초록색 삼각형으로 표시, 라벨은 virginica로 표시

plt.xlabel(df.columns[0])  #x축의 이름을 df행의 0번째 인덱스 sepal length(cm)로 표시
plt.ylabel(df.columns[1])  #y축의 이름을 df행의 1번째 인덱스 sepal width(cm)로 표시
plt.legend()  #라벨 표시
plt.show()    #표시
```

```python
# pteal 그래프

plt.plot(setosa.iloc[:,[2]], setosa.iloc[:,[3]], 'ro', label = "setosa")
# setosa의 전체 데이터의 전체 중 0번째 인덱스(petal length(cm))와 
# 전체 중 1번째 인덱스(pteal width) 를 통해 x, y축에 해당하는 값을 빨간 원으로 표시, 라벨은 setosa로 표시

plt.plot(versicolor.iloc[:,[2]], versicolor.iloc[:,[3]], 'bx', label = "versicolor")
# versicolor의 전체 데이터의 전체 중 0번째 인덱스(petal length(cm))와 
# 전체 중 1번째 인덱스(pteal width) 를 통해 x, y축에 해당하는 값을 파란 x로 표시, 라벨은 versicolor로 표시

plt.plot(virginica.iloc[:,[2]], virginica.iloc[:,[3]], 'g^', label = "virginica")
# virginica의 전체 데이터의 전체 중 0번째 인덱스(petal length(cm))와 
# 전체 중 1번째 인덱스(pteal width) 를 통해 x, y축에 해당하는 값을 초록색 삼각형으로 표시, 라벨은 virginica로 표시

plt.xlabel(df.columns[2])  #x축의 이름을 df행의 2번째 인덱스 petal length(cm)로 표시
plt.ylabel(df.columns[3])  #y축의 이름을 df행의 3번째 인덱스 pteal width(cm)로 표시
plt.legend()  #라벨 표시
plt.show()
```
