### 학생들 점수를 생성하는 함수

```python
def get_exel():            # DataFrame을 이용하여 가시성 증가된 exel함수
    Score = get_point()
    sc = pd.DataFrame(Score, index = ['No.1', 'No.2', 'No.3', 'No.4', 'No.5', 'No.6', 'No.7', 'No.8', 'No.9', 'No.10'])
    sc.columns = ['Korean', 'English', 'Math', 'Science']
    sc.index.name = 'No'                                     # 가시성을 위한가로 세로 정렬
                                        # 가시성을 위한가로 세로 정렬
    return(sc)
```

```python
def get_point():           # 가시성이 없는 점수만 있는 2차원 배열
    Score = np.random.randint(100, size=(10, 4)) # 4행 10열의 데이터 생성
    return(Score)
```

### 1번 문제

```python
def num1():    
    sc = get_exel()
    plt.scatter(sc['English'], sc['Korean']) # 가로를 영어, 세로를 수학으로 흩뿌림
    plt.xlabel('English')                    # x축을 영어
    plt.ylabel('Korean')                     # y축을 국어
```

### 2번 문제

```python
def num2():
    sc = get_exel()
    
    fig, axs = plt.subplots(2, 2, figsize=(20, 10))     # 2행 2열의 그림 총 4개로 구성, 사이즈는 20 x 10으로 구성

    axs[0, 0].bar(sc['Korean'], sc['Korean'], label = 'Korean')
    axs[0, 0].legend()                                  # 국어를 표시

    axs[0, 1].bar(sc['English'], sc['English'], label = 'English')
    axs[0, 1].legend()                                  # 영어를 표시     

    axs[1, 0].bar(sc['Math'], sc['Math'], label = 'Math')
    axs[1, 0].legend()                                  # 수학을 표시

    axs[1, 1].bar(sc['Science'], sc['Science'], label = 'Science')
    axs[1, 1].legend()                                  # 과학을 표시
```

### 3번 문제

```python
def num3():
    sc = get_exel()
    Score = get_point()
    forcir = Score[:, 0:1] # 국어 점수만 빼냄
    import itertools
    forcir2 = list(itertools.chain(*forcir)) #2차원 배열을 1차원 배열로 배치
    A = 0
    B = 0
    C = 0                                    # 비율을 계산하기 위해 각 학점을 초기화
    D = 0
    for i in range(10):
        if forcir2[i] >= 90:
            A = A+1
        elif 80 <= forcir2[i] < 90:          # 해당하는 점수에 있으면 추가, 
            B = B+1
        elif 70 <= forcir2[i] < 80:
            C = C+1
        elif forcir2[i] < 70:
            D = D+1
    values = [A, B, C, D]                    # 비율을 나타내는 values 배열 새로 생성
    # print(values)
    labels = ['A','B','C','D']               # ABCD 순서에 맞게 각 ABCD로 이름 나타냄
    plt.pie(values, labels=labels)           # 원그래프 생성
    plt.legend(loc=(1.2, 0.3))               # ABCD 가 각각 어떤 색깔을 표시하는지, 그리고 그 표시하는 것의 위치를 설정
    plt.show()
```

### 4번 문제

```python
def num4():
    sc = get_exel()
    
    # 최소, 중간, 최대 값을 갖는 각각의 리스트 생성
    Koxnm = [sc['Korean'].min(), sc['Korean'].median(), sc['Korean'].max()]     
    Engnm = [sc['English'].min(), sc['English'].median(), sc['English'].max()]
    Matnm = [sc['Math'].min(), sc['Math'].median(), sc['Math'].max()]
    Sicnm = [sc['Science'].min(), sc['Science'].median(), sc['Science'].max()]
    arrlist=([Koxnm, Engnm, Matnm, Sicnm])
    # 2차원 배열로 arrlist를 생성

    fig1, ax = plt.subplots(nrows=1,ncols=1, figsize=(8, 4))
    # 8*4의 사이즈를 갖고, 가로1, 세로 1을 갖음
    ax.boxplot(arrlist) 
    # 박스플롯 형태의 그래프 출력하는데, 각 값을 순서대로 국어, 영어, 수학, 과학으로 표시
    ax.set_xticklabels(['Korean', 'English', 'Math', 'Science'])
    plt.show() # 그림 출력
```
