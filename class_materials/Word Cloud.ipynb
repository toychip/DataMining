```python
plt.figure(figsize=(11,11)) #이미지 사이즈 지정
plt.imshow(wordcloud, interpolation='lanczos') #이미지의 부드럽기 정도
plt.axis('off') #x y 축 숫자 제거
plt.show() 


```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt 
 
stopwords = set(STOPWORDS) # 만들어놓고 리스트를 만들어서 넣어놈
stopwords.add('워드클라우드')  # Stopword 정의되어있는 라이브러리  
stopwords.add('ICT') 
stopwords.add('분석') 
 
#wordcloud = WordCloud(font_path='C:\Windows\Fonts\\HMFMMUEX.ttc',stopwords=stopwords,background_color='white').generate(text)
wordcloud = WordCloud(font_path='~/Library/Fonts/Arial Unicode.ttf',stopwords=stopwords,background_color='white').generate(text)

plt.figure(figsize=(11,11)) #이미지 사이즈 지정
plt.imshow(wordcloud, interpolation='lanczos') #이미지의 부드럽기 정도
plt.axis('off') #x y 축 숫자 제거
plt.show() 
```

```python
import numpy as np
from PIL import Image

text = open('./worldcloud/alice.txt').read()
alice_mask= np.array(Image.open('./worldcloud/alice.png'))

stopword=set(STOPWORDS)
stopword.add('said')
```

```python
plt.figure(figsize=(8,8))
plt.imshow(alice_mask,cmap=plt.cm.gray, interpolation='bilinear') 
# alice는 그냥 그리고, mask 이미지 읽은.부분에서 넘파이 어레이로 0101로 읽음
# mask로 그림을 그림
plt.axis('off')
plt.show
```

```python
wc = WordCloud(background_color='white',max_words=2000,mask=alice_mask, stopwords= stopword)
wc = wc.generate(text)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import numpy as np
from PIL import Image

#text = open('./data/alice.txt').read()
alice_coloring= np.array(Image.open('./worldcloud/alice_color.png'))

stopword=set(STOPWORDS)
stopword.add('said')

wc = WordCloud(background_color='white',max_words=2000,mask=alice_coloring, stopwords= stopwords)
wc = wc.generate(text)
```

```python
from wordcloud import ImageColorGenerator
image_colors=ImageColorGenerator(alice_coloring)
#image_colors.default_color = [0,0,0]
print(image_colors.image.shape)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(alice_coloring,cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func = image_colors), interpolation='bilinear')
# recolor는 원래의 색깔에 톤에 맞춰서 나타내줌
plt.axis('off')
plt.show()
```
